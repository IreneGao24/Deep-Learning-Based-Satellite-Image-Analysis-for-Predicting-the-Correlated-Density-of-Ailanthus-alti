```python
import os
import sys
import csv
import pandas as pd
import numpy as np

import os.path
from os import path

from skimage.color import rgb2hsv
import PIL
from PIL import Image


import deepforest
from deepforest import main

from pykml import parser

import geopandas as gpd 
import fiona 
from bs4 import BeautifulSoup
```

```python
img_paths = []
kml_paths = []
files = os.listdir("/home/irene/ISEF-Project-2023/Data/Density AM")
for f in files:
    if(f[-4:]=='.jpg'):
        img_paths.append(f)
    else:
        kml_paths.append(f)
kml_paths = sorted(kml_paths)
img_paths = sorted(img_paths)
```

```python
c = 0
index = []
u = []
l = []
for i in kml_paths:  
    with open('/home/irene/ISEF-Project-2023/Data/Density AM/'+i, 'r') as f:
        s = BeautifulSoup(f, 'xml')
        upper = ""
        lower = ""
        for node in s.select('north'):
            n = str(node)
            n = n[7:]
            n = n[:-8]
            print(n)
            upper = upper + n + ", "
        for node in s.select('south'):
            n = str(node)
            n = n[7:]
            n = n[:-8]
            print(n)
            lower = lower + n + ", "
        for node in s.select('east'):
            n = str(node)
            n = n[6:]
            n = n[:-7]
            print(n)
            lower = lower + n
        for node in s.select('west'):
            n = str(node)
            n = n[6:]
            n = n[:-7]
            print(n)
            upper = upper + n
        index.append("/home/irene/ISEF-Project-2023/Data/Density AM/"+i[:-3]+"jpg")
        u.append(upper)
        l.append(lower)
        print("/home/irene/ISEF-Project-2023/Data/Density AM/"+i[:-3]+"jpg")
        print(upper)
        print(lower)
        c = c+1
```

```python
d = {'Path':np.array(index), 'Upper':np.array(u), 'Lower':np.array(l)}
df = pd.DataFrame(data=d)
```
