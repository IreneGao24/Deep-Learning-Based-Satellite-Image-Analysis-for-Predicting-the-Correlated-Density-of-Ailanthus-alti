```python
am = pd.read_csv('/home/irene/ISEF-Project-2023/Data/AM.csv')
ll = pd.read_csv('/home/irene/ISEF-Project-2023/Data/Mappa_ailanto_Parco_Alta_Murgia_23.09.csv')
```

```python
Ones = []
Zeroes = []
for i in ll.index:
    x = ll["X"][i]
    y = ll["Y"][i]
    print(x)
    print(y)
    for j in am.index:
        u = am["Upper"][j].split(", ")
        l = am["Lower"][j].split(", ")
        if(float(y)<= float(u[0]) and float(y) >= float(l[0])):
            if(float(x) <= float(l[1]) and float(x) >= float(u[1])):
                if(Ones.count(am["Path"][j]) == 0):
                    Ones.append(am["Path"][j])
        else:
            if(Zeroes.count(am["Path"][j]) == 0):
                Zeroes.append(am["Path"][j])
```

```python
labels = [0]*len(Zeroes)
labels.extend([1]*len(Ones))
path_list = Zeroes+Ones
```

```python
df = pd.DataFrame({'image_path': path_list, 'label': labels})
df
```
